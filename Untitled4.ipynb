```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import folium
```

```python
df = pd.read_csv('/Users/Thor/Downloads/loc.csv')
df.head(5)
```

```python
# define the world map centered around Canada with a higher zoom level
world_map = folium.Map(location=[20.5937, 78.9629], zoom_start=4)


# display world map
#world_map
```

```python
# instantiate a feature group for the incidents in the dataframe
incidents = folium.map.FeatureGroup()

# loop through the 100 crimes and add each to the incidents feature group
for lat, lng, in zip(df.latitude, df.longitude):
    incidents.add_child(
        folium.CircleMarker(
            [lat, lng],
            radius=5, # define how big you want the circle markers to be
            color='yellow',
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
        )
    )

# add incidents to map
world_map
```

```python
from folium import plugins

# instantiate a mark cluster object for the incidents in the dataframe
incidents = plugins.MarkerCluster().add_to(world_map)

# loop through the dataframe and add each data point to the mark cluster
for lat, lng, label, in zip(df.latitude, df.longitude, df['name of centers']):
    folium.Marker(
        location=[lat, lng],
        icon=None,
        popup=label,
    ).add_to(incidents)

# display map
world_map
```
